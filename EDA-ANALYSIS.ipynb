```python
# Titanic EDA - CodeAlpha Internship
# Author: G Sanjay Srivastav

# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Dataset
df = pd.read_csv(r"C:\Users\sanja\Downloads\unzipped_folder/Titanic-Dataset.csv")
df.head()
```

```python
# Dataset info
df.info()

# Check missing values
df.isnull().sum()

# Describe dataset
df.describe()

# Value counts
df['Survived'].value_counts()
```

```python
# Q1. How many survived vs not?
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()

# Q2. Who survived more - Men or Women?
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()

# Q3. Survival by Passenger Class
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Class')
plt.show()
```
